```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= [[(i+j) for i in range(5)] for j in range(100)]
target=[(i+5) for i in range(100)]
```

```python
data=np.array(data, dtype=float)
target=np.array(target, dtype=float)
```

```python
data.shape
```

```python
x_train, x_test, y_train, y_test=train_test_split(data, target, test_size=0.2, random_state=4) 
```

```python
x_train = np.dstack(x_train).T
x_test = np.dstack(x_test).T
```

```python
x_train=x_train[..., newaxis]
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model=Sequential()

model.add(LSTM(units=1,
               return_sequences=True,
               batch_input_shape=(None, None, 1)))
model.add(LSTM(units=1, return_sequences=False))
```

```python
model.compile(optimizer='adam', loss='mean_absolute_error', metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit(x_train, y_train, validation_split=0.1, epochs=500)
```

```python
result=model.predict(x_test)
```

```python
plt.scatter(range(20), result, c='g')
plt.scatter(range(20), y_test, c='r')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.show()
```
