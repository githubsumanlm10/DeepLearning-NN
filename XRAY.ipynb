```
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dropout, Flatten, Dense, Reshape
from keras.optimizers import Adam
import os
import matplotlib.pyplot as plt
%matplotlib inline

from keras.callbacks import ModelCheckpoint
```

```
images = '/content/drive/My Drive/Diease/ChestXray/test'
```

```
img = ImageDataGenerator().flow_from_directory(images, target_size=(256,256), classes=['NORMAL', 'PNEUMONIA'],  batch_size=32,
                                             class_mode='categorical', shuffle=True)
```

```
len(img)
```

```
img_r = img[0][0].shape[1]
img_c = img[0][0].shape[2]
```

```
img_r
```

```
model = Sequential()
```

```
model.add(Conv2D(filters=128, kernel_size=(8,8), activation='relu', input_shape=(256,256, 3)))
model.add(Conv2D(filters=128, kernel_size=(8,8), activation='relu'))

model.add(MaxPool2D())

model.add(Conv2D(filters=64, kernel_size=(6,6), activation='relu'))
model.add(Conv2D(filters=64, kernel_size=(6,6), activation='relu'))

model.add(MaxPool2D())

model.add(Conv2D(filters=32, kernel_size=(4,4), activation='relu'))
model.add(Conv2D(filters=32, kernel_size=(4,4), activation='relu'))

model.add(MaxPool2D())

model.add(Conv2D(filters=16, kernel_size=(4,4), activation='relu'))

model.add(Flatten())

model.add(Dense(2, activation='softmax', ))


opt = Adam(0.001)
model.compile(optimizer=opt, loss="categorical_crossentropy", metrics=model.compile(optimizer=opt, loss="categorical_crossentropy", metrics=['accuracy']))

#  CHECKPOINTING MODEL

checkpoint_path = 'training_xry/cp.ckpt'

checkpoint = os.path.dirname(checkpoint_path)

cp_callback =ModelCheckpoint(checkpoint ,save_weights_only=True, verbose=1)
```

```
model.fit_generator(img, epochs=10, callbacks= [cp_callback] )
```

```
model.load_weights(checkpoint)
```
